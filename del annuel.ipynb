```python
import numpy as np
import pandas as pd
df=pd.read_excel("del_annuel.xlsx",engine="openpyxl")
d=df.iloc[9]
df
```

```python
index_with_nan = df.index[df.isnull().any(axis=1)]
index_with_nan
df.drop(index_with_nan,0, inplace=True)
df
```

```python
df.set_axis( d, axis='columns', inplace=True)
df
```

```python
bon_df=df.rename(columns={'RESULTAT': 'Total_Credits'})
bdf=bon_df.columns.fillna("Decision")
bon_df.columns=bdf
bon_df=bon_df.reset_index(drop=True)
bon_df
```

```python
tri_df=bon_df.sort_values(by = ["Total_Credits","Nom","Prénom"],
                ascending = [False,True,True])
tri_df
```

```python
g=bon_df.Decision.value_counts()
g
total_etudiant=len(tri_df)
pourc=round(g*100/total_etudiant,1)
pourc
f=pourc-round(pourc,0)
f.to_numpy
if sum(pourc)<100:
    pourc[np.argmax(f)]+=0.1
pourc
delib=[g,pourc]
np.array(delib)
delib=np.transpose(delib)
DF=pd.DataFrame(delib)
DF.set_axis(["nbre etudiant","pourcentage"],axis='columns',inplace=True)
DF.set_axis(g.index,axis="index",inplace=True)
SD=DF
DF
```

```python
import matplotlib.pyplot as plt
labels = DF.index
serie = DF['pourcentage']
separation = (0, 0, 0, 0) # Séparation des tranches
fig, ax = plt.subplots()
ax.pie(serie, explode=separation, labels=labels, autopct='%1.1f%%',
shadow=True, startangle=90)
ax.axis('equal')
plt.show()
```
