2. En se basant sur les tableaux des coefficients, écrire un programme qui cacule la moyenne
de chaque UE et détermine si l’étudiant l’a validée ou pas.


```python
#physique
phy=exam.iloc[:,2:5]
coef=phy['Physique_1']
coef2=phy['Physique_2']
coef_1=coef*2.5
coef_P2=coef2*2.5
coef_3=phy['TP_Physique']
g={"Physique1" : coef_1, "Physique2" : coef_P2, 
        "TPPhysique" : coef_3 }
df_phy=pd.DataFrame(g)
df_phy
moy_phy=(np.sum(df_phy,axis=1))/6
moy_phy
valide=[]
for i in range(len(moy_phy)):
    if moy_phy[i]>=10:
        valide.append('V')
    else:
        valide.append('NV')
```

```python
#MATH
math=exam.iloc[:,5:7]
coef_M=math['Maths_1']
coef_M2=math['Maths_2']
coef_M1=coef_M*2.5
coef_M2=coef_M2*2.5
M={"MATH1" : coef_M1, "MATH2" : coef_M2}
df_math=pd.DataFrame(M)
df_math
moy_math=(np.sum(df_math,axis=1))/5
moy_math
valide_M=[]
for i in range(len(moy_math)):
    if moy_math[i]>=10:
        valide_M.append('V')
    else:
        valide_M.append('NV')
```

```python
#CHIMIE
chimie=exam.iloc[:,7:10]
coef_C=chimie['Chimie1']
coef_C2=chimie['Chimie2']
coef_C1=coef_C*2.5
coef_C2=coef_C2*2.5
coef_C3=chimie['TPChimie']
C={"CHIMIE1" : coef_C1, "CHIMI2" : coef_C2, "TPCHIMIE" : coef_C3}
df_chimie=pd.DataFrame(C)
df_chimie
moy_chimie=(np.sum(df_chimie,axis=1))/6
moy_chimie
valide_C=[]
for i in range(len(moy_chimie)):
    if moy_chimie[i]>=10:
        valide_C.append('V')
    else:
        valide_C.append('NV')
```

```python
#informatique
inf=exam.iloc[:,10:12]
coef_I=inf['Info1']
coef_i=inf['Info2']
coef_I1=coef_I*2.5
coef_I2=coef_i*2.5
I={"INFO1" : coef_I1, "INFO2" : coef_I2 }
df_inf=pd.DataFrame(I)
df_inf
moy_inf=(np.sum(df_inf,axis=1))/5
moy_inf
valide_I=[]
for i in range(len(moy_inf)):
    if moy_inf[i]>=10:
        valide_I.append('V')
    else:
        valide_I.append('NV')
```

```python
#BASE
base=exam.iloc[:,12:15]
coef_B=base['Base1']
coef_b=base['Base2']
coef_B1=coef_B*1.5
coef_B2=coef_b*1.5
coef_B3=base['Base3']
B={"Base1" : coef_B1, "BASE2" : coef_B2, 
        "BASE3" : coef_B3 }
df_base=pd.DataFrame(B)
df_base
moy_base=(np.sum(df_base,axis=1))/4
moy_base
valide_B=[]
for i in range(len(moy_base)):
    if moy_base[i]>=10:
        valide_B.append('V')
    else:
        valide_B.append('NV')
```

```python
#GRAPHIQUE
graf=exam.iloc[:,15:17]
coef_G=graf['Graphique1']
coef_g=graf['Graphique2']
coef_G1=coef_G*2
coef_G2=coef_g*2
G={"Graphique1" : coef_G1, "Graphique2" : coef_G2 }
df_graf=pd.DataFrame(G)
df_graf
moy_graf=(np.sum(df_graf,axis=1))/4
moy_graf
valide_G=[]
for i in range(len(moy_graf)):
    if moy_graf[i]>=10:
        valide_G.append('V')
    else:
        valide_G.append('NV')
```

6-Écrire un programme qui détermine le major de la classe. Les crédits des UE sont respectivement 12, 10, 12, 10, 8 et 8.


```python
s=pd.read_excel("examens.xlsx",engine="openpyxl")
dr=s
dr.insert(5,"moy_Phy", moy_phy, allow_duplicates=False)
dr.insert(8,"moy_math",moy_math,allow_duplicates=False)
dr.insert(10,"moy_chimie",moy_chimie, allow_duplicates=False)
dr.insert(15,"moy_inf", moy_inf, allow_duplicates=False)
dr.insert(19,"moy_BASE", moy_base, allow_duplicates=False)
dr.insert(22,"moy_graph", moy_graf, allow_duplicates=False)
tri_moy=dr.sort_values(by = ["moy_Phy","moy_math","moy_chimie","moy_inf","moy_BASE","moy_graph"],
                ascending = [False,True,True,True,True,True],na_position='first')
tri_moy
D=tri_moy.iloc[0]
maj_prenom=D[1]
maj_nom=D[0]
print("major de la classe est:",maj_prenom,maj_nom)
```

7. Écrire un programme qui détermine pour chaque UE, le nombre d’étudiants validés. Les
résultats obtenus seront rangés dans une liste.

```python
ph=dr['moy_Phy']
v_phy=0
for i in range(len(ph)):
    if ph[i]>10:
        v_phy+=1
    else:
        v_phy+=0
M=dr['moy_math']
v_math=0
for i in range(len(M)):
    if M[i]>10:
        v_math+=1
    else:
        v_math+=0
C=dr['moy_chimie']
v_chimie=0
for i in range(len(C)):
    if C[i]>10:
        v_chimie+=1
    else:
        v_chimie+=0
IF=dr['moy_inf']
v_inf=0
for i in range(len(IF)):
    if IF[i]>10:
        v_inf+=1
    else:
        v_inf+=0
Bs=dr['moy_BASE']
v_base=0
for i in range(len(Bs)):
    if Bs[i]>10:
        v_base+=1
    else:
        v_base+=0
GF=dr['moy_graph']
v_graf=0
for i in range(len(GF)):
    if GF[i]>10:
        v_graf+=1
    else:
        v_graf+=0
valid=[v_phy,v_math,v_chimie,v_inf,v_base,v_graf]
print("nbr d'etudiant validées par UE :",valid)
```

```python
x = ['phy','math','chimie','inf','base','graph']
y = valid
fig ,ax=plt.subplots(figsize=(10,5))
plt.plot(x, y, label="validée en f(UE)")
plt.grid()
plt.show() 
```
